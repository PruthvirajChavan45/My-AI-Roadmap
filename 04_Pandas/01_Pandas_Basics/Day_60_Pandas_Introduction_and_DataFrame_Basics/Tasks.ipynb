##**ðŸ”¹ Task 1: Create DataFrame**

Create a Pandas DataFrame named df using a dictionary with this data:

| Name   | Age | City      |
| ------ | --- | --------- |
| Rahul  | 22  | Pune      |
| Anita  | 25  | Mumbai    |
| Suresh | 30  | Delhi     |
| Neha   | 28  | Bangalore |


````python
# Your Code Here :

import pandas as pd

data = {
    "Name" : ["Rahul","Anita","Suresh","Neha"],
    "Age" : [22,25,30,28],
    "City" : ["Pune","Mumbai","Delhi","Bangalore"]
}

df = pd.DataFrame(data)
df
````

##**ðŸ”¹ Task 2: Inspect the DataFrame**

Without modifying the DataFrame:

1. Display first 3 rows

2. Display last 2 rows

3. Show info()

4. Print:
  *   shape
  *   dtypes


````python
# Your Code Here :

import pandas as pd

# Display first 3 rows
print("First 3 Rows : \n",df.head(3))

# Display last 2 rows
print("\nLast 2 Rows : \n",df.tail(2))

# Show info()
print("\n Information : \n")
df.info()


# Print :
# 1) shape :
print("\nShape : \n",df.shape)

# 2) dtypes :
print("\n Data Types : \n",df.dtypes)
````

##**ðŸ”¹ Task 3 Requirements**

Using the same DataFrame df:

1. Select only the Name column
2. Select Name and City columns together

````python
# Select only the Name column
select_col = df["Name"]
print("Selecting Name Column : \n",select_col)

# Select Name and City columns together
multiple_col = df[["Name","City"]]
print("\nSelecting Name and City Columns : \n",multiple_col)
````

##**ðŸ”¹ Task 4: Row Selection**

1. Select the second row using .iloc

2. Select the row where index label = 1 using .loc

(Observe carefully if output is same or different)

````python
# Select the second row using .iloc
select_row = df.iloc[1]
print("Second Row using .iloc : \n",select_row)

# Select the row where index label = 1 using .loc
label = df.loc[1]
print("\nIndex Label using .loc : \n",label)

# Both are same
````
