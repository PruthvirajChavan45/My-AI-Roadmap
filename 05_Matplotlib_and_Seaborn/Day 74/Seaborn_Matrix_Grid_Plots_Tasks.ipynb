
# Seaborn â€“ Matrix / Grid Plots

Matrix and grid plots are widely used in Exploratory Data Analysis (EDA)
to analyze relationships between multiple numerical variables.

These plots use color intensity to represent values and help identify
patterns, correlations, and feature groupings before modeling.



## 1. Import Libraries


```python

import matplotlib.pyplot as plt
import seaborn as sns
```


## 2. Load Dataset

The Iris dataset is used because it contains multiple numerical features
that are ideal for correlation and matrix-based analysis.


```python

iris = sns.load_dataset("iris")
iris.head()
```


## 3. Correlation Matrix

A correlation matrix shows the strength and direction of relationships
between numerical variables.


```python

corr = iris.corr(numeric_only=True)
corr
```


## 4. Basic Heatmap

Heatmap visualizes correlation values using color gradients.


```python

sns.heatmap(corr)
plt.title("Correlation Between Numerical Features")
plt.show()
```


## 5. Heatmap with Annotations

Annotations help display correlation values directly inside cells.


```python

sns.heatmap(corr, annot=True)
plt.title("Annotated Correlation Heatmap")
plt.show()
```


## 6. Heatmap with Color Map

Color maps improve interpretation of positive and negative correlations.


```python

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap with Color Map")
plt.show()
```


## 7. Heatmap Styling

Styling improves readability and presentation quality.


```python

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Styled Correlation Heatmap")
plt.show()
```


## 8. Cluster Map

Clustermap applies hierarchical clustering to group similar features.


```python

sns.clustermap(corr, annot=True)
```


## 9. Cluster Map with Color Map

Color maps improve cluster visualization.


```python

sns.clustermap(corr, cmap="viridis")
```


## 10. Heatmap & Clustermap Summary

- Correlation values close to **+1** indicate strong positive relationships.
- Correlation values close to **-1** indicate strong negative relationships.
- Values near **0** represent weak or no correlation.
- Heatmaps visualize correlation strength using color intensity.
- Clustermaps group similar features together using hierarchical clustering.



## 11. Final Conclusion

Matrix and grid plots play a vital role in data science workflows.

They help in:
- understanding feature relationships
- detecting multicollinearity
- selecting relevant features
- identifying patterns before modeling

These visualizations are essential before applying machine learning models.
