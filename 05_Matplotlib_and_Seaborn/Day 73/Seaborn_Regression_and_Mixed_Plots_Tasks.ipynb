
# Seaborn â€“ Regression & Mixed Plots

This notebook focuses on regression and mixed plots in Seaborn.
These plots are widely used in Exploratory Data Analysis (EDA) to
understand relationships, trends, and correlations between variables.

Regression and mixed plots combine scatter plots, statistical models,
and distribution plots to provide deeper analytical insights.



## 1. Import Libraries


```python

import matplotlib.pyplot as plt
import seaborn as sns
```


## 2. Load Dataset

The **tips** dataset is used to analyze regression relationships
between total bill and tip amount.


```python

tips = sns.load_dataset("tips")
tips.head()
```


## 3. Regression Plot (regplot)

Regplot displays a scatter plot along with a regression line.
It is useful for identifying linear relationships between variables.


```python

sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Relationship Between Total Bill and Tip")
plt.show()
```


## 4. Regression Line Only

Scatter points can be hidden to visualize only the regression trend.
This is useful when focusing purely on relationship direction.


```python

sns.regplot(data=tips, x="total_bill", y="tip", scatter=False)
plt.title("Regression Line Between Total Bill and Tip")
plt.show()
```


## 5. Polynomial Regression

Polynomial regression helps visualize non-linear relationships.
Here, a second-order polynomial is used.


```python

sns.regplot(data=tips, x="total_bill", y="tip", order=2)
plt.title("Polynomial Regression (Order = 2)")
plt.show()
```


## 6. lmplot with Hue

lmplot supports grouping and multiple regression lines.
Different regression trends are compared using the `hue` parameter.


```python

sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```


## 7. lmplot with Faceting

Faceting allows automatic creation of multiple regression plots
based on categorical variables.


```python

sns.lmplot(data=tips, x="total_bill", y="tip", col="time")
plt.show()
```


## 8. Jointplot (Scatter)

Jointplot combines a central scatter plot with marginal distributions
for both variables.


```python

sns.jointplot(data=tips, x="total_bill", y="tip")
plt.show()
```


## 9. Jointplot with Regression

This plot combines joint distribution with a regression line.


```python

sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
plt.show()
```


## 10. Jointplot with KDE

KDE jointplot shows density-based relationship between variables.


```python

sns.jointplot(data=tips, x="total_bill", y="tip", kind="kde")
plt.show()
```


## 11. Pairplot (Basic)

Pairplot provides an overview of relationships between multiple numerical features.


```python

iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```


## 12. Pairplot with Hue

Hue allows category-wise separation and comparison of feature relationships.


```python

sns.pairplot(iris, hue="species")
plt.show()
```


## 13. Pairplot Summary

- Pairplot is commonly used as the initial step of EDA.
- It provides a visual overview of relationships among numerical features.
- Diagonal plots represent feature distributions.
- The `diag_kind="kde"` option can be used to show density plots.
- Hue helps visualize class-wise separation.



## 14. Final Conclusion

Regression and mixed plots help in understanding:

- linear and non-linear relationships
- correlation patterns
- feature interactions
- data structure before modeling

These plots play a critical role in feature selection and model building
during data science workflows.
