
# Seaborn â€“ Distribution Plots

This notebook focuses on understanding data distribution using Seaborn.
Distribution plots help analyze how numerical values are spread, identify
skewness, density, and compare patterns across categories.

These plots play a crucial role in Exploratory Data Analysis (EDA)
before applying statistical or machine learning models.



## 1. Import Libraries

In this section, required Python libraries for visualization are imported.


```python

import matplotlib.pyplot as plt
import seaborn as sns
```


## 2. Load Dataset

The Iris dataset is used for distribution analysis because it contains
multiple numerical features suitable for histogram and density plots.


```python

iris = sns.load_dataset("iris")
iris.head()
```


## 3. Understanding Distribution

Distribution helps us understand how values are spread within a dataset.

It provides insight into:
- frequency of values
- data spread
- skewness
- overall shape of numerical features



## 4. Basic Histogram

Histogram shows the frequency distribution of numerical values
by dividing the data into bins.


```python

sns.histplot(data=iris, x="sepal_length")
plt.title("Distribution of Sepal Length")
plt.show()
```


## 5. Histogram with Custom Bins

Changing the number of bins helps analyze how bin size
affects interpretation of data distribution.


```python

sns.histplot(data=iris, x="sepal_length", bins=20)
plt.title("Distribution of Sepal Length with Custom Bins")
plt.show()
```


## 6. Histogram with KDE

Combining histogram with KDE allows visualization of both
frequency and smooth density curve together.


```python

sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Histogram with KDE")
plt.show()
```


## 7. KDE Plot

KDE plot represents probability density and provides a smooth
curve to understand distribution shape and skewness.


```python

sns.kdeplot(data=iris, x="sepal_length", fill=True)
plt.title("KDE Distribution of Sepal Length")
plt.show()
```


## 8. Rug Plot

Rug plot displays individual data points along the axis
and helps observe exact value distribution.


```python

sns.rugplot(data=iris, x="sepal_length")
plt.title("Rug Plot of Sepal Length")
plt.show()
```


## 9. Multiple Distributions using Hue (Histogram)

Using hue allows comparison of distributions across different categories.


```python

sns.histplot(data=iris, x="sepal_length", hue="species")
plt.title("Distribution of Sepal Length by Species")
plt.show()
```


## 10. Multiple Distributions using Hue (KDE)

KDE with hue provides smooth comparison of distribution shapes
across categories.


```python

sns.kdeplot(data=iris, x="sepal_length", hue="species", fill=True)
plt.title("KDE Distribution by Species")
plt.show()
```


## 11. ECDF Plot

ECDF (Empirical Cumulative Distribution Function) shows the cumulative
percentage of observations below a given value.


```python

sns.ecdfplot(data=iris, x="sepal_length")
plt.title("ECDF of Sepal Length")
plt.show()
```


## 12. Professional Distribution Plot

Histogram combined with KDE is widely used in industry-level
EDA to analyze numerical features.


```python

sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Professional Distribution Plot")
plt.show()
```


## 13. Distribution Summary

Key observations from distribution analysis:

- Histogram shows frequency distribution of numerical values.
- KDE plot represents probability density and data shape.
- Distplot is deprecated and replaced by histplot and kdeplot.
- Rug plot displays individual observations.
- Hue enables category-wise distribution comparison.
- ECDF plot shows cumulative probability distribution.



## 14. Final Conclusion

Distribution analysis helps understand data behavior before modeling.

This notebook demonstrates:
- data spread and skewness
- comparison between categories
- importance of distribution plots in EDA

These insights are essential for feature engineering,
outlier detection, and model selection in data science projects.
