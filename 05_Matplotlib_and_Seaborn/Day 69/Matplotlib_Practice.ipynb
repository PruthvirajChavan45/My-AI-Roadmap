## ðŸ§ª Task 1: Basic Line Plot

This task demonstrates how to create a simple line plot using Matplotlib pyplot.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## ðŸ§ª Task 2: Figure and Canvas

This task shows how to control figure size and resolution using figsize and dpi.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.figure(figsize=(8, 4), dpi=100)
plt.plot(x, y, marker='o', color='purple')
plt.title('Canvas Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## ðŸ§ª Task 3: Line Styles and Markers

This task demonstrates customization of line styles, markers, colors, and legends.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

plt.figure(figsize=(8, 4))

plt.plot(x, [1, 2, 3, 4, 5], '--o', color='red', label='Line 1')
plt.plot(x, [2, 4, 6, 8, 10], ':^', color='blue', label='Line 2')
plt.plot(x, [5, 4, 3, 2, 1], '-s', color='green', label='Line 3')

plt.title('Line Styles and Markers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## ðŸ§ª Task 4: Subplots using pyplot

This task explains how to create multiple plots in one figure using pyplot.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], marker='o')
plt.title('Squares')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot([1, 2, 3, 4], [1, 8, 27, 64], marker='o')
plt.title('Cubes')
plt.grid(True)

plt.show()
```

## ðŸ§ª Task 5: Object-Oriented Subplots

This task demonstrates Matplotlib's Object-Oriented interface.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, [1, 4, 9, 16], marker='o')
ax[0].set_title('Squares')
ax[0].grid(True)

ax[1].plot(x, [1, 8, 27, 64], marker='o')
ax[1].set_title('Cubes')
ax[1].grid(True)

plt.show()
```

## ðŸ§ª Task 6: Legends, Labels and Titles

This task shows how to use legends and labels for better readability.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [40, 30, 20, 10]

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y1, marker='o', label='Increasing')
ax.plot(x, y2, marker='o', label='Decreasing')

ax.set_title('Multiple Lines')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.legend()
ax.grid(True)

plt.show()
```

## ðŸ§ª Task 7: Axis Limits

This task demonstrates how to control visible data range.

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 5, 10, 15, 20, 25]

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y, marker='o')
ax.set_xlim(0, 6)
ax.set_ylim(0, 30)
ax.set_title('Axis Limits Example')
ax.grid(True)

plt.show()
```

## ðŸ§ª Task 8: Scatter Plot

This task visualizes the relationship between two numerical variables.

```python
import matplotlib.pyplot as plt

study_hours = [5, 7, 8, 10, 12, 15]
marks = [50, 60, 65, 80, 85, 100]

fig, ax = plt.subplots(figsize=(8, 4))

ax.scatter(study_hours, marks, s=80)
ax.set_title('Study Hours vs Marks')
ax.set_xlabel('Study Hours')
ax.set_ylabel('Marks')
ax.grid(True)

plt.show()
```

## ðŸ§ª Task 9: Histogram

This task explains how to visualize data distribution.

```python
import matplotlib.pyplot as plt

scores = [45, 50, 55, 60, 60, 65, 70, 72, 75, 80, 85, 90, 95]

fig, ax = plt.subplots(figsize=(7, 4))

ax.hist(scores, bins=5, edgecolor='black')
ax.set_title('Score Distribution')
ax.set_xlabel('Scores')
ax.set_ylabel('Frequency')

plt.show()
```

## ðŸ§ª Task 10: Boxplot

This task demonstrates outlier detection using boxplots.

```python
import matplotlib.pyplot as plt

marks = [35, 40, 45, 50, 55, 60, 65, 70, 90]

fig, ax = plt.subplots(figsize=(8, 4))

ax.boxplot(marks, patch_artist=True,
           boxprops=dict(facecolor='lightblue'))

ax.set_title('Distribution of Marks')
ax.set_ylabel('Marks')

plt.show()
```
