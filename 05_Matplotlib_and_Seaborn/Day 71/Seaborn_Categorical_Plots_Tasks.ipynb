
# Seaborn Categorical Plots

This notebook demonstrates the use of Seaborn categorical plots for
exploratory data analysis (EDA).

Categorical plots are used when one or more variables are categorical,
such as gender, day, or time. These plots help compare distributions,
averages, and frequencies across categories.



## Dataset

We use the built-in **tips** dataset from Seaborn.

The dataset contains information about restaurant transactions, including:
- total bill amount
- tip amount
- gender
- day of visit
- time (lunch/dinner)
- table size


```python

import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```


## 1. Countplot

Countplot is used to show the frequency of categorical values.
It helps understand how often each category appears in the dataset.


```python

sns.countplot(data=tips, x="sex")
plt.title("Customer Count by Gender")
plt.show()
```


## 2. Countplot with Hue

Using hue allows comparison of category counts across another variable.
Here, we compare daily visits by gender.


```python

sns.countplot(data=tips, x="day", hue="sex")
plt.title("Customer Count by Day and Gender")
plt.show()
```


## 3. Barplot

Barplot displays the average (mean) of a numerical variable for each category.
This is useful for comparing group-wise averages.


```python

sns.barplot(data=tips, x="sex", y="total_bill")
plt.title("Average Total Bill by Gender")
plt.show()
```


## 4. Barplot with Hue

This plot compares average values across multiple categorical groups.


```python

sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Average Total Bill by Day and Gender")
plt.show()
```


## 5. Boxplot

Boxplots help visualize data distribution, spread, median,
and potential outliers.


```python

sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Distribution of Total Bill Across Days")
plt.show()
```


## 6. Violinplot

Violin plots combine boxplot statistics with distribution density,
providing deeper insight into data spread.


```python

sns.violinplot(data=tips, x="day", y="total_bill")
plt.title("Distribution Shape of Total Bill by Day")
plt.show()
```


## 7. Stripplot

Stripplots display individual observations.
They are useful for small datasets but may show overlapping points.


```python

sns.stripplot(data=tips, x="day", y="total_bill")
plt.title("Individual Total Bill Observations by Day")
plt.show()
```


## 8. Swarmplot

Swarmplots improve stripplots by adjusting points to avoid overlap,
making patterns easier to observe.


```python

sns.swarmplot(data=tips, x="day", y="total_bill")
plt.title("Non-overlapping Total Bill Observations by Day")
plt.show()
```


## 9. Boxplot with Hue

This plot compares distributions across multiple categories,
helping identify differences between groups.


```python

sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Distribution of Total Bill by Day and Gender")
plt.show()
```


## 10. Violinplot with Hue

Violinplot with hue allows comparison of distribution shapes
between different categories.


```python

sns.violinplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Distribution Shape of Total Bill by Day and Gender")
plt.show()
```


## Summary

In this notebook, Seaborn categorical plots were used to analyze customer behavior.

Key learnings:
- countplot helps analyze frequency
- barplot compares average values
- boxplot identifies spread and outliers
- violinplot shows full distribution
- stripplot and swarmplot visualize raw data
- hue enables multi-category comparison

These plots form a critical part of exploratory data analysis (EDA)
in data science and analytics.
