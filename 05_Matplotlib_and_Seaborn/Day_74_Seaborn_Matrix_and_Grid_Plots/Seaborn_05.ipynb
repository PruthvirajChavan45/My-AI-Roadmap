## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


### Heatmaps (heatmap())

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

# Correlation heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Cluster Map (clustermap())

```python
# Clustered heatmap
sns.clustermap(corr, annot=True, cmap="viridis")
```

### Styling & Themes

```python
# Change global style
sns.set_style("darkgrid")       # options: white, dark, darkgrid, whitegrid, ticks

# Change color palette
sns.set_palette("Set2")

# Apply to a plot
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=tips, s=80)
plt.title("Styled Scatter with Darkgrid & Palette")
plt.show()
```
