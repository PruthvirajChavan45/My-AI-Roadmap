This notebook focuses on Seaborn relational plots including scatter plots and line plots.
It also covers styling and color palettes used during Exploratory Data Analysis (EDA).

The goal of this notebook is to understand how relationships between variables
can be visualized using Seaborn.



## Dataset Used

We use the built-in **tips** dataset provided by Seaborn.

The dataset contains information about restaurant bills including:
- total bill amount
- tip amount
- gender
- smoking status
- day and time
- table size


```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```


## Task 1: Basic Scatter Plot

A scatter plot is used to visualize the relationship between two numerical variables.
Here, we analyze the relationship between **total bill** and **tip amount**.


```python

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Total Bill vs Tip")
plt.show()
```


## Task 2: Scatter Plot with Hue

The `hue` parameter is used to group data based on a categorical variable.
Different colors represent different categories.

Here, we group data by **gender**.


```python

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Total Bill vs Tip by Gender")
plt.show()
```


## Task 3: Scatter Plot with Hue and Style

We can visualize multiple categorical variables together.
- `hue` controls color grouping
- `style` controls marker grouping

Here, we use gender and time.


```python

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")
plt.title("Total Bill vs Tip by Gender and Time")
plt.show()
```


## Task 4: Scatter Plot with Size

Bubble plots allow visualization of an additional numerical variable.
The size of each marker represents the table size.


```python

sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
plt.title("Total Bill vs Tip with Table Size")
plt.show()
```


## Task 5: Line Plot

Line plots are used to analyze trends in ordered data.
Seaborn automatically calculates mean values and displays confidence intervals.


```python

sns.lineplot(data=tips, x="size", y="total_bill")
plt.title("Table Size vs Total Bill")
plt.show()
```


## Task 6: Line Plot with Hue

Line plots can compare trends across different categories.
Here, we compare spending patterns between genders.


```python

sns.lineplot(data=tips, x="size", y="total_bill", hue="sex")
plt.title("Table Size vs Total Bill by Gender")
plt.show()
```


## Task 7: Line Plot without Confidence Interval

Confidence intervals are shown by default in Seaborn.
They can be removed for cleaner visualization.


```python

sns.lineplot(data=tips, x="size", y="total_bill", errorbar=None)
plt.title("Line Plot without Confidence Interval")
plt.show()
```


## Task 8: Seaborn Styles

Seaborn provides different visual themes to change plot appearance.
These styles help improve readability and presentation.


```python

sns.set_style("whitegrid")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Whitegrid Style Example")
plt.show()
```


## Task 9: Color Palettes

Color palettes control the colors used in plots.
They are especially useful when visualizing categorical variables.


```python

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    palette="Set2"
)
plt.title("Scatter Plot with Set2 Color Palette")
plt.show()
```
