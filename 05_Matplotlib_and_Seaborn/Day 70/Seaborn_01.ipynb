#  Seaborn – Introduction & Relational Plots


## What is Seaborn?

Seaborn is a **high-level data visualization library** built on top of Matplotlib.  
It provides:
- Beautiful, modern default styles  
- Concise syntax (less code for complex visuals)  
- Integration with Pandas DataFrames  

Perfect for quick, attractive plots while keeping full Matplotlib power underneath.


###  What Are Relational Plots?

Relational plots are used to **show the relationship between two variables**.

- They focus on how one variable changes with respect to another.
- Typically used for continuous (numeric) data, but can also mix with categories.

In Seaborn, the main relational plots are:

1. **`scatterplot()`** → displays individual data points  
2. **`lineplot()`** → shows a continuous trend or summary line

> Both functions belong to Seaborn’s *relational* family.


## You can load sample DataSets using SNS

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Built-in sample datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
penguins = sns.load_dataset("penguins")

print(tips.head())
print(iris.head())
print(penguins.head())
```

### Scatter Plot with scatterplot()

```python
# Scatter: Total bill vs Tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

### Line Plot with lineplot()

```python
# Line plot: day vs tip (average per day)
sns.lineplot(x="day", y="tip", data=tips, estimator="mean", marker="o")
plt.title("Line Plot: Average Tip by Day")
plt.show()
```

### Styling & Color Palettes

```python
# Change global style
sns.set_style("darkgrid")        # options: white, dark, darkgrid, whitegrid, ticks

# Try different color palettes
sns.set_palette("deep")         # others: deep, bright, dark, muted, colorblind

sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips, s=80)
plt.title("Styled Scatter with Hue & Palette")
plt.show()
```
