游릭 Task 4: Aggregate Functions (Overall)
---



Using matrix m, print:

1. Sum of all elements

2. Mean of all elements

3. Maximum value

```python
print("Sum of All Elements : ",np.sum(m))
print("Mean of All Elements : ",np.mean(m))
print("Maximum Value : ",np.max(m))
```

游릭 Task 5: Aggregate Functions (Axis-based)
---



 Using matrix m, print:

1. Sum across rows

2. Sum across columns

```python
print("Sum of Rows : ", np.sum(m, axis=1))
print("Sum of Columns : ", np.sum(m, axis=0))
```
