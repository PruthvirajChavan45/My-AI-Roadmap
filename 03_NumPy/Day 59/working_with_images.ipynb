```python
# Plotting 3d graphs
# Meshgrids

import numpy as np

a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
b
```

```python
xx,yy = np.meshgrid(a,b)
xx
```

```python
import matplotlib.pyplot as plt
plt.scatter(xx,yy)
```

```python
def func(x,y):
  return x**2 + y**2

zz = func(xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()

fig.add_trace(go.Surface(x = xx, y = yy, z = zz))
fig.show()
```
