```python
import numpy as np

arr = np.array([[10, 20, 30], [40, 50, 60]])
print(arr)
```

游릭 Task 1: Array Attributes
---



Using arr, print:

* shape

* ndim

* size

* dtype

```python
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

游릭 Task 2: Indexing
---

From arr, extract:

* Element 50

* Entire first row

* Entire second column

 Print each result.

```python
print(arr[1,1])
print(arr[0])
print(arr[:,1])
```

游릭 Task 3: Slicing
---



From arr, slice and print:

* Sub-array containing:

      20 30
      50 60

(Think rows + columns range)

```python
print(arr[:,1:])
```

游릭 Task 4: Boolean Indexing
---

Using arr, extract all elements greater than 30.

 Print the result.

```python
print(arr[arr > 30])
```

游릭 Task 5: Reshaping
---


Reshape arr into a 1D array of length 6.

Print the reshaped array.

```python
print(arr.reshape(6))
```
