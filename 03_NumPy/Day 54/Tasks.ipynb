```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
```

游릭 Task 1: Fancy Indexing
---

From arr, extract elements at indices:

      0, 2, 4

Print the result.

```python
print("Extract Elements From Perticular Indices : ",arr[[0,2,4]])
```

游릭 Task 2: Conditional Selection
---

From arr, extract only values greater than 25.

Print the result.

```python
print("Greater Than 25 Values are : ", arr[arr > 25])
```

游릭 Task 3: Sorting (1D)
---

 Sort arr in ascending order

```python
print("Sorted Array : ",np.sort(arr))
```

游릭 Task 4: Ordering (Argsort)
---

Get the indices that would sort arr

```python
print("Indices of That Array Are : ",np.argsort(arr))
```

游릭 Task 5: Concatenation (1D)
---



Now move forward

Create:

    a = np.array([1, 2, 3])
    b = np.array([4, 5, 6])

Concatenate a and b into a single array

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("After Concatenate Array Look Like : ", np.concatenate([a,b]))
```

游릭 Task 6: Stacking (2D)
---

Using the same arrays a and b:

1.  Stack them vertically

2. Stack them horizontally

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Vertically Stack : \n", np.vstack((a,b)))
print("\n Horizontally Stack : \n", np.hstack((a,b)))
```
